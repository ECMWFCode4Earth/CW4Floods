```python
import pandas as pd
from cw.cfg import DATA_DIR
import numpy as np
from cw.misc import str_to_num, haversine_dist
```

```python
station_data = pd.read_csv(DATA_DIR + "station_with_efas_v1.csv")
stream_info = pd.read_csv(DATA_DIR + "stream_info.csv", delimiter=";")

merged_df = station_data.merge(stream_info, on="ROOT_ID")

merged_df.WATER_LEVEL = merged_df.WATER_LEVEL.map(str_to_num)

merged_df.to_csv(DATA_DIR + "ml_data_v1.csv", index = False) 
```

```python
lookup_table = pd.read_csv(DATA_DIR + "EFAS_stations_for_esowc.csv")
```

```python
dist_threshold = 2.5 #KM
station_mini = set()
for i, rows in merged_df.iterrows():
    lat_1 = rows.LATITUDE
    lon_1 = rows.LONGITUDE
    for j, (lat_2, lon_2) in enumerate(zip(lookup_table.StationLat.values, lookup_table.StationLon.values)):
        c1 = (lat_1,lon_1)
        c2 = (lat_2, lon_2)
        dist = haversine_dist(c1, c2)
        if dist<dist_threshold:
            station_mini.add((rows.ROOT_ID, lookup_table.ObsID[j], dist))
station_mini
```
